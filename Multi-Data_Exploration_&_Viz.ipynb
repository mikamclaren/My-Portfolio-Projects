```python
# Import dependencies and magics
import pandas as pd
import json
import numpy as np
import matplotlib as mpl
import math
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Load the data file to a data frame named df. 
df = pd.read_csv('datacombined.csv')
df
```

```python
# describe the series in the dataframe
df.describe()
```

```python
# Count of the x columns

for i in range (1,5):
    col = 'x'+str(i)
    print(f'Count {col},{len(df[col])}')
```

```python
# Mean of the x columns to two decimal places
import statistics
mean_x1 = statistics.mean(df['x1'])
mean_x2 = statistics.mean(df['x2'])
mean_x3 = statistics.mean(df['x3'])
mean_x4 = statistics.mean(df['x4'])

print(f'mean_x1: {mean_x1:.2f}\tmean_x2: {mean_x2:.2f}')
print(f'mean_x3: {mean_x3:.2f}\tmean_x4: {mean_x4:.2f}')
```

```python
# Mean of the y columns to two decimal places

mean_y1 = statistics.mean(df['y1'])
mean_y2 = statistics.mean(df['y2'])
mean_y3 = statistics.mean(df['y3'])
mean_y4 = statistics.mean(df['y4'])

print(f'mean_y1: {mean_y1:.2f}\tmean_y2: {mean_y2:.2f}')
print(f'mean_y3: {mean_y3:.2f}\tmean_y4: {mean_y4:.2f}')
```

```python
# Sample variance of the x columns to two decimal places

svar_x1 = df['x1'].var() 
svar_x2 = df['x2'].var() 
svar_x3 = df['x3'].var()
svar_x4 = df['x4'].var() 

print(f'svar_x1: {svar_x1:.2f}\tsvar_x2: {mean_x2:.2f}')
print(f'svar_x3: {svar_x3:.2f}\tsvar_x4: {svar_x4:.2f}')
```

```python
# Sample variance of the y columns to two decimal places

svar_y1 = df['y1'].var() 
svar_y2 = df['y2'].var() 
svar_y3 = df['y3'].var()
svar_y4 = df['y4'].var() 

print(f'svar_y1: {svar_y1:.2f}\tsvar_y2: {svar_y2:.2f}')
print(f'svar_y3: {svar_y3:.2f}\tsvar_y4: {svar_y4:.2f}')
```

```python
# Calculate min and max for axis base of data in dataframe

# Did you mean 'Calculate min and max for axis based on data in dataframe'? note sure what's needed here

x_bounds = df[['x1','x2','x3','x4']].max(axis=1).max(axis=0)
y_bounds = df[['y1','y2','y3','y4']].max(axis=1).max(axis=0)

# Set a style

plt.style.use('bmh')

# Plot the scatter diagrams

x1 = df['x1'].tolist()
y1 = df['y1'].tolist()
x2 = df['x2'].tolist()
y2 = df['y2'].tolist()
x3 = df['x3'].tolist()
y3 = df['y3'].tolist()
x4 = df['x4'].tolist()
y4 = df['y4'].tolist()

plt.scatter(x1, y1, color= 'red', marker='x', label='x1y1')
plt.scatter(x2, y2, color= 'green', marker='^', label='x2y2')
plt.scatter(x3, y3, color= 'orange', marker='*', label='x3y3')
plt.scatter(x4, y4, color= 'purple', marker='.', label='x4y4')

# title - at top of chart

plt.title('Datacombined.csv')

# axis labels

plt.xlabel('x values')
plt.ylabel('y values')

# add legend

plt.legend()

# set axis limits

plt.xlim(0, x_bounds * 1.1)
plt.ylim(0, y_bounds * 1.1)
```

```python
# Create a new figure with four identically sized subplots in a quadrant 


plt.style.use('default')
fig, ax = plt.subplots(2, 2, sharex='col', sharey='row')
                    

# Add super title

fig.suptitle("Anscombe's Quartet by Shemeika McLaren", fontsize=14)

# Adjust axis limits

ax[0,0].plot(df['x1'], df['yr1'], linestyle='-', color='k')
ax[0,1].plot(df['x2'], df['yr2'], linestyle='-', color='k')
ax[1,0].plot(df['x3'], df['yr3'], linestyle='-', color='k')
ax[1,1].plot(df['x4'], df['yr4'], linestyle='-', color='k')

# First plots

ax[0,0].axis(xmin=0,xmax=22)
ax[0,0].axis(ymin=0,ymax=14)
ax[0,0].text(5, 10, 'data1')
ax1 = ax[0,0].scatter(x1, y1, color= 'red', marker='x')

# Second plots

ax[0,1].axis(xmin=0,xmax=22)
ax[0,1].axis(ymin=0,ymax=14)
ax[0,1].text(5, 10, 'data2')
ax2 = ax[0,1].scatter(x2, y2, color= 'green', marker='^')

# Third plots

ax[1,0].axis(xmin=0,xmax=22)
ax[1,0].axis(ymin=0,ymax=14)
ax[1,0].text(5, 10, 'data3')
ax3 = ax[1,0].scatter(x3, y3, color= 'orange', marker='*')

# Fourth plots

ax[1,1].axis(xmin=0,xmax=22)
ax[1,1].axis(ymin=0,ymax=14)
ax[1,1].text(5, 10, 'data4')
ax4 = ax[1,1].scatter(x4, y4, color= 'purple', marker='.')


plt.savefig('Anscombe_McLaren.png')
```
